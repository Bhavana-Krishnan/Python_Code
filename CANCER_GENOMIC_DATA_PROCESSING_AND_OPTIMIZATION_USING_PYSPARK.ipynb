Merging Patient and sample data

```python
import pandas as pd
```

```python
df1 = pd.read_excel('Clinical_patient_sample_data.xlsx',sheet_name='Clinical_patient_data 1')
df2 = pd.read_excel('Clinical_patient_sample_data.xlsx',sheet_name='Clinical_sample_data 1')
```

```python
m = {'left_only': 'df1', 'right_only': 'df2', 'both': 'df1, df2'}
df1 = df1.reset_index(drop = True)
result = df1.merge(df2, on=[df2['PATIENT_ID'].str[:-3]], how='outer', indicator='B')
result['B'] = result['B'].map(m)
c = open("/content/concate_brca 1.txt", 'w')
e = result.to_string()
c. write(e)
c.close()
```
