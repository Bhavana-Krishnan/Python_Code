```python
import itertools
import numpy as np
import codecs
```

```python
x=[]
for v in itertools.product(['A', 'T', 'G','C'], repeat=9):x.append("".join(list(map(str,v))))
```

```python
x
```

```python
A=[]
T=[]
G=[]
C=[]
for j in range(len(x[0])):
    A.append([i[j] for i in x].count('A'))
    T.append([i[j] for i in x].count('T'))
    G.append([i[j] for i in x].count('G'))
    C.append([i[j] for i in x].count('C'))
print ("A", A)
print ("T", T)
print ("G", G)
print ("C", C)
```

```python
n = [A,T,G,C]
n = np.array(n)
print(n)
percentages = n/n.sum(axis=0) * 100
print(percentages)
```
