Extracting ALLELEID values from the VCF file

```python

   
# Lets use a regular expression to match a date string 
regex = r'ALLELEID=[0-9]+'
file = open('clinvar_20220227_10K.vcf','r')
res_1=[]

for i in file:
  match = re.findall(regex, i)
  if match == []:
    res_1.append('-')
  else:
    y = ' '.join([str(elem) for elem in match])
    res_1.append(y)
print(len(res_1))
res1=[]
for i in res_1:
  if i ==' ':
    res_1.append('-')
  else:
    x = i.replace('ALLELEID=','')
    res1.append(x)
res1[0] ='ALLELEID'
print(len(res1))
print(res1)
file.close()
```

Extracting CLNHGVS values from the VCF file

```python

regex = r'CLNHGVS=NC_000001.10:g.[0-9]+[ATGC]>[ATGC]'
file = open('clinvar_20220227_10K.vcf','r')
l=[]
for i in file:
  match = re.findall(regex, i )
  l.append(match)
res2 = [''.join(ele) for ele in l]
res2=[i.replace('CLNHGVS=','') for i in res2]
res2[0] = 'CLNHGVS'
print(len(res2))
'''file1 = open('/content/drive/MyDrive/ALLELEID.csv','w')
res2=[]
for k in file1:
  res2.append(k)
for r, s in zip(res2, res):
  file1.write('{}\n'.format(k))
file1.close()    '''
file.close()
```

Extracting CLNSIG values from the VCF file

```python

regex = r'CLNSIG=(Uncertain_significance|Likely_benign|Benign)'
file = open('clinvar_20220227_10K.vcf','r')
l=[]
for i in file:
  match = re.findall(regex, i )
  l.append(match)
print(l[0:10])
res3 = [''.join(ele) for ele in l]
res3=[i.replace('CLNSIG=','') for i in res3]
res3[0] = 'CLNSIG'
print(res3)
file.close()
```

Extracting CLNVC values from the VCF file

```python

regex = r'CLNVC=(Deletion|Duplication|Indel|Insertion|Inversion|Microsatellite|single_nucleotide_variant)'
file = open('clinvar_20220227_10K.vcf','r')
l=[]
for i in file:
  match = re.findall(regex, i )
  l.append(match)
print(l[0:10])
res4= [''.join(ele) for ele in l]
res4=[i.replace('CLNVC=','') for i in res4]
res4[0] = 'CLNVC'
print(res4)
file.close()
```

Extracting ORIGIN values from the VCF file

```python

regex = r'ORIGIN=[0-9]*'
file = open('clinvar_20220227_10K.vcf','r')
res_5=[]
for i in file:
  match = re.findall(regex, i)

  if match == []:
    res_5.append('-')
  else:
    y = ' '.join([str(elem) for elem in match])
    res_5.append(y)
print(len(res_5))
res5=[]
for i in res_5:
  x = i.replace('ORIGIN=','')
  res5.append(x)
res5[0] ='ORIGIN'
print(len(res5))
print(res5)
file.close()
```

Extracting RS values from the VCF file

```python

regex = r'RS=[0-9]*'
file = open('clinvar_20220227_10K.vcf','r')
res6=[]
for i in file:
  match = re.findall(regex, i)

  if match == []:
    res6.append('-')
  else:
    y = ' '.join([str(elem) for elem in match])
    x = y.replace('RS=','')
    res6.append(x)
print(len(res6))

res6[0] ='RS'

print(res6)
file.close()
```

Extracting GENE_SYMBOL and GENE_ID values from the VCF file

```python

regex = r'GENEINFO=[A-Z]*:[0-9]*'
file = open('clinvar_20220227_10K.vcf','r')
res_7=[]
gene_sym = []
gene_id = []
for i in file:
  match = re.findall(regex, i)

  if match == []:
    res_7.append(' ')
  else:
    y = ' '.join([str(elem) for elem in match])
    res_7.append(y)
print(len(res_7),res_7)

for c in res_7:
  if c == ' ':
    gene_sym.append('-')
    gene_id.append('-')
  else:
    y1, y2 = c.split(':')   
    z = y1.replace('GENEINFO=','')
    gene_sym.append(z)
    gene_id.append(y2) 
#print(len(res6))
#res6[0] ='RS'
gene_sym[0] = 'GENE_SYMBOL'
gene_id[0] = 'GENE_ID'
print(len(gene_sym), gene_sym)
print(len(gene_id),gene_id)
```

Extracting CONSEQUENCE values from the VCF file

```python

regex = r'\|[a-z]+_[a-z]+'
file = open('clinvar_20220227_10K.vcf','r')
res_8 = []
Conseq = []

for i in file:
  match = re.findall(regex, i)

  if match == []:
    res_8.append('-')
  else:
    y = ' '.join([str(elem) for elem in match])
    res_8.append(y)
print(res_8)
Conseq=[]
for i in res_8:
  if c == '-':
    Conseq.append('-')
  else:
#    i = i + '\n'
    x = i.replace('|','')
    Conseq.append(x)
Conseq[0] = 'Consequence'

print(Conseq)
file.close()
```
